```python
import os
import re
import shutil

df2 = "/kaggle/input/datasets/saitejaswinichikoti/medicinal-leaf-dataset/Medicinal Leaf Dataset/Segmented Medicinal Leaf Images"
df2_output = "/kaggle/working/Medicinal_Leaf_Dataset"

if os.path.exists(df2_output):
    shutil.rmtree(df2_output)
os.makedirs(df2_output)

def standardize_name(name):
    name = re.sub(r'\(.*?\)', '', name).strip()
    return name.replace("-", "_").replace(" ", "_").lower().strip()

scientific_names_df2 = []

for label in os.listdir(df2):
    label_path = os.path.join(df2, label)
    if os.path.isdir(label_path):
        scientific_name = standardize_name(label)
        scientific_names_df2.append(scientific_name)
        dest_path = os.path.join(df2_output, scientific_name)
        shutil.copytree(label_path, dest_path)
        print(f"Copied: {label} --> {scientific_name}")

scientific_names_df2 = list(set(scientific_names_df2))
print("\nFinal scientific names list df2:")
print(scientific_names_df2)
```

```python
df4 = "/kaggle/input/datasets/saitejaswinichikoti/south-indian-medicinal-plant-dataset/SIMPD V1 South Indian Medicinal Plants dataset (Version 1)"
df4_output = "/kaggle/working/SIMPD_Dataset"

manual_fixes = {
    "ouretlanata": "ouret_lanata",
    "eclipta_prostrate": "eclipta_prostrata",
    "justica_adhatoda": "justicia_adhatoda",
    "senna_atriculata": "senna_auriculata",
    "eichhornia_crassipes": "eichhornia_crassipes",
    "hibiscus_rosasinensis": "hibiscus_rosa_sinensis",
    "aloe_barbadensis_miller": "aloe_barbadensis",
}

exclude_from_df4 = {"hibiscus_rosa_sinensis", "calotropis_gigantea", "senna_auriculata"}

def standardize_name(name):
    std = name.replace("-", "_").replace(" ", "_").lower().strip()
    if std in manual_fixes:
        std = manual_fixes[std]
    return std

if os.path.exists(df4_output):
    shutil.rmtree(df4_output)
os.makedirs(df4_output)

scientific_names_df4 = []

for label in os.listdir(df4):
    label_path = os.path.join(df4, label)
    if os.path.isdir(label_path):
        scientific_name = standardize_name(label)
        
        if scientific_name in exclude_from_df4:
            print(f"Skipping: {label} --> {scientific_name}")
            continue
        
        scientific_names_df4.append(scientific_name)
        dest_path = os.path.join(df4_output, scientific_name)
        shutil.copytree(label_path, dest_path)
        print(f"Copied: {label} --> {scientific_name}")

scientific_names_df4 = list(set(scientific_names_df4))
print("\nFinal scientific names list df4:")
print(scientific_names_df4)
```

```python
set_df2 = set(scientific_names_df2)
set_df3 = set(scientific_names_df3)
set_df4 = set(scientific_names_df4)

# Individual sizes
print("Size of df2:", len(set_df2))
print("Size of df3:", len(set_df3))
print("Size of df4:", len(set_df4))

# Common in all three
common_all = set_df2 & set_df3 & set_df4
print("\nCommon in ALL three:", len(common_all), sorted(common_all))

# Common in pairs ( common only between pairs not including the other one )
common_df2_df3 = set_df2 & set_df3 - set_df4
print("\nCommon in df2 & df3 only:", len(common_df2_df3), sorted(common_df2_df3))

common_df2_df4 = set_df2 & set_df4 - set_df3
print("\nCommon in df2 & df4 only:", len(common_df2_df4), sorted(common_df2_df4))

common_df3_df4 = set_df3 & set_df4 - set_df2
print("\nCommon in df3 & df4 only:", len(common_df3_df4), sorted(common_df3_df4))

# Union
union_all = set_df2 | set_df3 | set_df4
print("\nTotal unique after merging:", len(union_all))
print("All classes:", sorted(union_all))
```

```python
import os
import shutil

df2_output = "/kaggle/working/Medicinal_Leaf_Dataset"
df3_output = "/kaggle/working/Mendeley_Dataset"
df4_output = "/kaggle/working/SIMPD_Dataset"
merged_output = "/kaggle/working/Merged_Dataset"

if os.path.exists(merged_output):
    shutil.rmtree(merged_output)
os.makedirs(merged_output)

def merge_folder_symlink(src, dest):
    os.makedirs(dest, exist_ok=True)
    for img in os.listdir(src):
        src_img = os.path.join(src, img)
        dst_img = os.path.join(dest, img)
        if os.path.isfile(src_img):
            if os.path.exists(dst_img):
                base, ext = os.path.splitext(img)
                counter = 1
                while os.path.exists(dst_img):
                    dst_img = os.path.join(dest, f"{base}_{counter}{ext}")
                    counter += 1
            os.symlink(src_img, dst_img)

for source in [df2_output, df3_output, df4_output]:
    for folder in os.listdir(source):
        folder_path = os.path.join(source, folder)
        if os.path.isdir(folder_path):
            dest_path = os.path.join(merged_output, folder)
            merge_folder_symlink(folder_path, dest_path)

final_folders = sorted(os.listdir(merged_output))
print(f"Total classes: {len(final_folders)}")
for f in final_folders:
    count = len(os.listdir(os.path.join(merged_output, f)))
    print(f"{f}: {count} images")
```
