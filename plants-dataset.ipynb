```python
import os
import re
import shutil

# =====================================================================
# 1. DEFINE PATHS AND INPUTS
# =====================================================================
df2_input = "/kaggle/input/datasets/saitejaswinichikoti/medicinal-leaf-dataset/Medicinal Leaf Dataset/Segmented Medicinal Leaf Images"
df3_input = "/kaggle/input/datasets/saitejaswinichikoti/mendeley/Indian Medicinal Leaves Image Datasets/Medicinal Leaf dataset"
df4_input = "/kaggle/input/datasets/saitejaswinichikoti/south-indian-medicinal-plant-dataset/SIMPD V1 South Indian Medicinal Plants dataset (Version 1)"

merged_output = "/kaggle/working/Merged_Dataset"

# Fresh start for the output directory
if os.path.exists(merged_output):
    shutil.rmtree(merged_output)
os.makedirs(merged_output)

# =====================================================================
# 2. HELPER FUNCTIONS
# =====================================================================

# Safe file copier that appends unique prefixes/counters to prevent overwrites
def safe_copy_file(src_img_path, dest_dir, prefix=""):
    os.makedirs(dest_dir, exist_ok=True)
    img_name = os.path.basename(src_img_path)
    base, ext = os.path.splitext(img_name)
    
    # Prepend dataset identifier to avoid cross-dataset collisions
    if prefix:
        img_name = f"{prefix}_{img_name}"
        base = f"{prefix}_{base}"
        
    dest_img_path = os.path.join(dest_dir, img_name)
    
    # Handle duplicate file names internally via an incremental counter
    counter = 1
    while os.path.exists(dest_img_path):
        dest_img_path = os.path.join(dest_dir, f"{base}_{counter}{ext}")
        counter += 1
        
    shutil.copy2(src_img_path, dest_img_path)

# =====================================================================
# 3. PROCESSING DF2 (Medicinal Leaf Dataset)
# =====================================================================
print("Processing DF2...")

def standardize_df2(name):
    name = re.sub(r'\(.*?\)', '', name).strip()
    return name.replace("-", "_").replace(" ", "_").lower().strip()

for label in os.listdir(df2_input):
    label_path = os.path.join(df2_input, label)
    if os.path.isdir(label_path):
        scientific_name = standardize_df2(label)
        dest_class_dir = os.path.join(merged_output, scientific_name)
        
        for img in os.listdir(label_path):
            src_file = os.path.join(label_path, img)
            if os.path.isfile(src_file):
                safe_copy_file(src_file, dest_class_dir, prefix="df2")

# =====================================================================
# 4. PROCESSING DF3 (Mendeley Dataset)
# =====================================================================
print("\nProcessing DF3...")

mapping_df3 = {
    "Tecoma": "Tecoma stans", "Malabar_Nut": "justicia_adhatoda", "Curry": "Murraya koenigii",
    "Drumstick": "Moringa oleifera", "Tomato": "Solanum lycopersicum", "Lemongrass": "Cymbopogon citratus",
    "Ginger": "Zingiber officinale", "Chilly": "Capsicum frutescens", "Nelavembu": "Andrographis paniculata",
    "Raddish": "Raphanus sativus", "Nooni": "Morinda citrifolia", "Jackfruit": "Artocarpus heterophyllus",
    "Turmeric": "Curcuma longa", "Palak(Spinach)": "Spinacia oleracea", "Honge": "Millettia pinnata",
    "Mint": "Mentha", "Pumpkin": "Cucurbita moschata", "Lantana": "Lantana camara", "Jasmine": "Jasminum",
    "Castor": "Ricinus communis", "Papaya": "Carica papaya", "Sapota": "Manilkara zapota", "Kambajala": "Cassia tora",
    "Neem": "Azadirachta indica", "Guava": "Psidium guajava", "Bringaraja": "Eclipta prostrata",
    "kamakasturi": "Ocimum basilicum", "Coriender": "Coriandrum sativum", "Coffee": "Coffea",
    "ashoka": "Saraca asoca", "Tulsi": "Ocimum tenuiflorum", "Pea": "Pisum", "Seethapala": "Annona reticulata",
    "Caricature": "Graptophyllum pictum", "Citron lime (herelikai)": "Citrus medica", "Catharanthus": "Catharanthus roseus",
    "Astma_weed": "Euphorbia hirta", "Pepper": "Piper nigrum", "Balloon_Vine": "Cardiospermum halicacabum",
    "Hibiscus": "Hibiscus rosa-sinensis", "Bhrami": "Bacopa monnieri", "Pomoegranate": "Punica granatum",
    "Aloevera": "Aloe barbadensis", "Eucalyptus": "Eucalyptus globulus", "Beans": "Phaseolus",
    "Insulin": "Chamaecostus cuspidatus", "camphor": "Cinnamomum camphora", "Padri": "Radermachera xylocarpa",
    "Parijatha": "Nyctanthes arbor-tristis", "Marigold": "Tagetes erecta", "Taro": "Colocasia esculenta",
    "Spinach1": "Spinacia oleracea", "Bamboo": "Bambusoideae", "Amla": "Phyllanthus emblica",
    "Arali": "Ficus religiosa", "kepala": "Ixora undulata", "Globe Amarnath": "Gomphrena globosa",
    "Ganigale": "Calotropis gigantea", "Lemon": "Citrus limon", "Common rue(naagdalli)": "Ruta graveolens",
    "Ekka": "Calotropis gigantea", "Betel": "Piper betle", "Tamarind": "Tamarindus indica", "Henna": "Lawsonia inermis",
    "Gasagase": "Papaver somniferum", "Doddpathre": "Coleus amboinicus", "Malabar_Spinach": "Basella alba",
    "Rose": "Rosa", "Amruthaballi": "Tinospora cordifolia", "Kohlrabi": "Brassica oleracea var. gongylodes",
    "Nerale": "Syzygium cumini", "Mango": "Mangifera indica", "Onion": "Allium cepa", "Thumbe": "Leucas aspera",
    "Sampige": "Magnolia champaca", "Ganike": "Solanum nigrum",
}

for label in os.listdir(df3_input):
    label_path = os.path.join(df3_input, label)
    if os.path.isdir(label_path):
        if label in mapping_df3:
            # Resolves uppercase and spaces directly from dictionary mappings
            scientific_name = mapping_df3[label].replace("-", "_").replace(" ", "_").lower().strip()
            dest_class_dir = os.path.join(merged_output, scientific_name)
            
            for img in os.listdir(label_path):
                src_file = os.path.join(label_path, img)
                if os.path.isfile(src_file):
                    safe_copy_file(src_file, dest_class_dir, prefix="df3")

# =====================================================================
# 5. PROCESSING DF4 (SIMPD Dataset)
# =====================================================================
print("\nProcessing DF4...")

manual_fixes_df4 = {
    "ouretlanata": "ouret_lanata", "eclipta_prostrate": "eclipta_prostrata", "justica_adhatoda": "justicia_adhatoda",
    "senna_atriculata": "senna_auriculata", "eichhornia_crassipes": "eichhornia_crassipes",
    "hibiscus_rosasinensis": "hibiscus_rosa_sinensis", "aloe_barbadensis_miller": "aloe_barbadensis",
}
exclude_from_df4 = {"hibiscus_rosa_sinensis", "calotropis_gigantea", "senna_auriculata"}

for label in os.listdir(df4_input):
    label_path = os.path.join(df4_input, label)
    if os.path.isdir(label_path):
        scientific_name = label.replace("-", "_").replace(" ", "_").lower().strip()
        
        if scientific_name in manual_fixes_df4:
            scientific_name = manual_fixes_df4[scientific_name]
            
        if scientific_name in exclude_from_df4:
            continue
            
        dest_class_dir = os.path.join(merged_output, scientific_name)
        
        for img in os.listdir(label_path):
            src_file = os.path.join(label_path, img)
            if os.path.isfile(src_file):
                safe_copy_file(src_file, dest_class_dir, prefix="df4")

print("\nSuccess! All datasets merged into /kaggle/working/Merged_Dataset.")
```
